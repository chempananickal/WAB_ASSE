# Alternate way to find the reverse dependencies ranking

Use in case Libraries.io isn't working out for some reason. 

## Prerequisites

- Have Python 3.14 installed
- Create a virtual environment and activate it (recommended)
- Install everything in requirements.txt
- Set up Google Cloud credentials for BigQuery access (see [BigQuery documentation](https://docs.cloud.google.com/python/docs/reference/bigquery/latest))
- Alternatively, copy and paste the code into a Jupyter notebook within Kaggle (after setting up an account and enabling BigQuery API access)

## Getting started

Once everything is set up, run the following code block. It should show the top 20 most depended upon Python packages.

```python
from google.cloud import bigquery
import pandas as pd

client = bigquery.Client()

query = """
WITH Dependencies AS (
SELECT name,
       REGEXP_EXTRACT(requirement, r"([^\s^\[^>^=]+)\s?\[?") dependency
FROM `bigquery-public-data.pypi.distribution_metadata`
CROSS JOIN UNNEST(requires_dist) AS requirement)
SELECT dependency package,
       COUNT(DISTINCT name) num_dependent_pip_packages
FROM Dependencies
GROUP BY package
ORDER BY num_dependent_pip_packages DESC
"""

df = client.query(query).result().to_dataframe(create_bqstorage_client=False)
df.head(10)
```
